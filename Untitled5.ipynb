## Data Loading Process

This section of the code is responsible for initializing the DataFrame and iterating over each text file in the specified directory. We perform the following steps:

- Set the directory path where the lyrics text files are stored.
- Initialize a DataFrame with columns 'Filename' and 'Lyrics'.
- Begin processing each file in the directory.
- For each text file:
  - Construct the file's full path.
  - Open the file and read its contents.
  - Append a new row to the DataFrame with the file's name and its lyrics.
- Finally, display the first few entries of the DataFrame to verify the data loading process.

```python
import pandas as pd
import os

# Updated folder path
folder_path = '/Users/gillian/Desktop/CDA5/Lyrics'
df = pd.DataFrame(columns=['Filename', 'Lyrics'])

print("Start processing folder:", folder_path)

for filename in os.listdir(folder_path):
    if filename.endswith('.txt') and not filename.startswith('.'):
        print("Processing text file:", filename)
        file_path = os.path.join(folder_path, filename)
        with open(file_path, 'r', encoding='utf-8') as file:
            lyrics = file.read()
            df = df.append({'Filename': filename, 'Lyrics': lyrics}, ignore_index=True)

print(df.head())
```

```python
def preprocess_text(text):
    return text.lower()

df['Processed_Lyrics'] = df['Lyrics'].apply(preprocess_text)

print(df.head())
```

### Natural Language Processing Pipeline
Here we define a function `nlp_pipeline` to process the text using spaCy's NLP capabilities. This function takes a string of text as input and performs the following tasks:
- **Tokenization**: Splitting the text into individual words or tokens.
- **Lemmatization**: Converting words to their base or dictionary form.
- **Part-of-Speech Tagging**: Identifying the grammatical parts of speech for each word.

```python
import spacy
nlp = spacy.load('en_core_web_sm')

def nlp_pipeline(text):
    doc = nlp(text)
    return {
        'Tokens': [token.text for token in doc],
        'Lemmas': [token.lemma_ for token in doc],
        'POS': [token.pos_ for token in doc]
    }

df[['Tokens', 'Lemmas', 'POS']] = df['Processed_Lyrics'].apply(lambda x: pd.Series(nlp_pipeline(x)))

print(df.head())
```

### Filtering Stop Words
In this section, we enhance the stop words list by adding context-specific words that are not informative for our analysis, such as contractions and interjections. The purpose is to filter out these additional non-informative words from our word frequency analysis to focus on more meaningful content.

```python
from nltk.corpus import stopwords
from collections import Counter
import nltk
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

additional_stopwords = {"i'm", "can't", "yeah", "no"}
stop_words = stop_words.union(additional_stopwords)

all_lyrics = ' '.join(df['Processed_Lyrics'])

words = all_lyrics.split()
filtered_words = [word for word in words if word.lower() not in stop_words]

word_counts = Counter(filtered_words)

most_common_filtered_words = word_counts.most_common(10)
print(most_common_filtered_words)
```

## Sentiment Analysis with TextBlob

In this section, we perform sentiment analysis on the processed lyrics. TextBlob is a library that offers various text processing services, including sentiment analysis.

```python
!pip install textblob
```

```python
from textblob import TextBlob

df['Sentiment'] = df['Processed_Lyrics'].apply(lambda x: TextBlob(x).sentiment.polarity)

plt.hist(df['Sentiment'], bins=20, color='blue')
plt.title('Sentiment Distribution')
plt.xlabel('Sentiment Polarity')
plt.ylabel('Frequency')
plt.show()
```
